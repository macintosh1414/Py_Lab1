```python
import csv
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime

# 1. Функція для отримання масиву значень стовпця
def extract_column_as_array(file_path, column_name):
    with open(file_path, 'r') as file:
        reader = csv.DictReader(file)
        return np.array([row[column_name] for row in reader], dtype=str)

# Загрузка даних
file_path = 'russia_losses_equipment.csv'

# 2. Створення масиву зі значень стовпця «tank» (накопичена кількість)
tank_data = extract_column_as_array(file_path, 'tank').astype(int)
dates = extract_column_as_array(file_path, 'date')

# 3. Обчислення добових втрат танків (різниця між поточним і попереднім значенням)
daily_tank_losses = np.diff(tank_data[::-1], prepend=0)  # переворачиваем данные для правильного обчислення

# Приводимо всі від'ємні значення до 0, тому що втрати не можуть бути від'ємними
daily_tank_losses[daily_tank_losses < 0] = 0

# 4. Знаходження 4 найбільших добових втрат танків
largest_losses = np.sort(daily_tank_losses)[-4:]

# 5. Кількість танків, знищених навесні 2023 року
spring_start = datetime.strptime('2023-03-01', '%Y-%m-%d')
spring_end = datetime.strptime('2023-05-31', '%Y-%m-%d')
spring_indices = [
    i for i, date in enumerate(dates)
    if spring_start <= datetime.strptime(date, '%Y-%m-%d') <= spring_end
]

# Збираємо тільки позитивні втрати для весняного періоду
spring_2023_tank_losses = daily_tank_losses[spring_indices]
spring_2023_tank_losses = spring_2023_tank_losses[spring_2023_tank_losses > 0].sum()

# 6. Середнє значення знищених танків за останні 100 днів
last_100_days_losses = daily_tank_losses[-100:].mean()

# 7. Графік втрат танків за минулий рік
last_year_indices = range(len(dates) - 365, len(dates))
last_year_tank_losses = daily_tank_losses[last_year_indices]

plt.figure(figsize=(8, 16), dpi=100)
plt.plot(
    np.arange(1, len(last_year_tank_losses) + 1),
    last_year_tank_losses,
    linestyle='--',
    color='red',
    label='Tank Losses'
)
plt.title('Tank Losses Over the Last Year')
plt.xlabel('Day')
plt.ylabel('Number of Tanks Lost')
plt.grid(color='gray', linestyle=':', linewidth=0.5)
plt.legend()
plt.savefig('tank_losses_last_year.png')
plt.show()

# Виведення результатів
print("Добові втрати танків:", daily_tank_losses)
print("4 найбільші добові втрати танків:", largest_losses)
print("Кількість знищених танків навесні 2023 року:", spring_2023_tank_losses)
print("Середнє значення втрат за останні 100 днів:", last_100_days_losses)
```
